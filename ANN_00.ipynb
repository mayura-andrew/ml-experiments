```python
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
```

```python
# 1. Read the data in CSV file
df = pd.read_csv("Mobile_Price_Classification-220531-204702.csv")

# Separate the features (X) and the target variable (y)
X = df.drop('price_range', axis=1)
y = df['price_range']

# 2. Split the data set
# 75% for training, 25% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Standardizing the data (Highly recommended for ANNs so they converge properly)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 3. ANN Model Architecture
model = Sequential()

# defining the input shpae explicitly
model.add(Input(shape=(X_train_scaled.shape[1],)))

# 1st hidden layer with 8 neurons/nodes
model.add(Dense(8, activation='relu'))

# 2nd hidden layer with 4 neurons/nodes
model.add(Dense(4, activation='relu'))

#Output layer with 1 node (since it's binary classification: high(1) and low(0))
model.add(Dense(1, activation='sigmoid'))

# 4. Compile the model
# Using 'binary_crossentropy' as the loss function since we are predicting 0 or 1
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("Starting model training...")

# 5. Fit and train the model for 100 epochs with a batch size of 32
history = model.fit(
    X_train_scaled,
    y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_test_scaled, y_test),
    verbose=1
)

# Evaluate the model to see how it performs on the unseen test data
loss, accuracy = model.evaluate(X_test_scaled, y_test, verbose=0)
print(f"\nFinal Test Accuracy: {accuracy*100:.2f}%")

# 6. Save the weights file
# Keras requires the weights file to end in '.weights.h5'
weights_filename = 'mobile_price_model.weights.h5'
model.save_weights(weights_filename)
print(f"Model weights saved successfully to '{weights_filename}'")
```
