```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score


df = pd.read_csv('w2_position_salaries-220925-153437.csv')
X = df[['Level']]
y = df['Salary']


print(df.head())
print(X)
print(y)
```

```python
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Predict and compute R2 Score
y_pred_lin = lin_reg.predict(X)
r2_lin = r2_score(y, y_pred_lin)
print(f"Simple Linear Regression R2: {r2_lin:.4f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x=X['Level'], y=y, label='Actual Salary')
sns.lineplot(x=X['Level'], y=y_pred_lin, color='red', label='Simple Linear Regression')
plt.title('Level vs Salary with Simple Linear Regression')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ----------------------------------------------------
# 5. Train Polynomial Regression (Degree 2, 3, and 4)
# ----------------------------------------------------

# --- Degree 2 ---
poly_reg_2 = PolynomialFeatures(degree=2)
X_poly_2 = poly_reg_2.fit_transform(X)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly_2, y)
y_pred_poly_2 = lin_reg_2.predict(X_poly_2)
r2_poly_2 = r2_score(y, y_pred_poly_2)
print(f"Polynomial Regression (Degree 2) R2: {r2_poly_2:.4f}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X['Level'], y=y, label='Actual Salary')
sns.lineplot(x=X['Level'], y=y_pred_poly_2, color='green', label='Polynomial Regression (Degree 2)')
plt.title('Level vs Salary with Polynomial Regression (Degree 2)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# --- Degree 3 ---
poly_reg_3 = PolynomialFeatures(degree=3)
X_poly_3 = poly_reg_3.fit_transform(X)

lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly_3, y)
y_pred_poly_3 = lin_reg_3.predict(X_poly_3)
r2_poly_3 = r2_score(y, y_pred_poly_3)
print(f"Polynomial Regression (Degree 3) R2: {r2_poly_3:.4f}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X['Level'], y=y, label='Actual Salary')
sns.lineplot(x=X['Level'], y=y_pred_poly_3, color='blue', label='Polynomial Regression (Degree 3)')
plt.title('Level vs Salary with Polynomial Regression (Degree 3)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# --- Degree 4 ---
poly_reg_4 = PolynomialFeatures(degree=4)
X_poly_4 = poly_reg_4.fit_transform(X)

lin_reg_4 = LinearRegression()
lin_reg_4.fit(X_poly_4, y)
y_pred_poly_4 = lin_reg_4.predict(X_poly_4)
r2_poly_4 = r2_score(y, y_pred_poly_4)
print(f"Polynomial Regression (Degree 4) R2: {r2_poly_4:.4f}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X['Level'], y=y, label='Actual Salary')
sns.lineplot(x=X['Level'], y=y_pred_poly_4, color='blue', label='Polynomial Regression (Degree 4)')
plt.title('Level vs Salary with Polynomial Regression (Degree 4)')
plt.xlabel('Level')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```
